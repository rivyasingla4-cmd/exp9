```python
#Declaration of Numpy
import numpy as np

a = np.array([1, 10, 20, 30, 40, 50, 500])
b = np.array([[1, 2, 3],[4, 5, 6]])

#Dimension of array
print(a.ndim)
print(b.ndim)

print(a)
print(b)


# Shape of array
print(a.shape)
print(b.shape)

print(a.dtype)
print(b.dtype)

#In-built functions using Numpy
np.zeros((2,3))

#identity matrix
print(np.eye(3))

print(np.arange(1,10,2))

#Generates evenly spaced
#step = stop - start / (num - 1)
print(np.linspace(0,1,4))

#addition and multiplication
print(b * 2)
print(a + 5)

print(np.mean(a))

print(np.median(a))

print(np.max(a))

print(np.min(a))

print(np.sum(a))
```
